```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.morphology import (erosion, dilation, closing, opening,
                                area_closing, area_opening)
from skimage.color import rgb2gray
```

```python
img = cv.imread('/content/drive/MyDrive/TI3D/PCVK/Modul9/j.png')
kernel = np.ones((5,5), np.uint8)
dilation = cv.dilate(img, kernel, iterations = 1)
plt.subplot(121),plt.imshow(img),plt.title('Citra Awal')
plt.subplot(122),plt.imshow(dilation),plt.title('Citra Hasil Dilasi')
```

```python
# Library
kernel = np.ones((7,7), np.uint8)
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(opening), plt.title('Citra Hasil Opening')
```

```python
kernel = np.ones((7,7), np.uint8)
#Function for erosion
def erosion(img, SE):
  imgErode= cv.erode(img,SE,1)
  return imgErode
#Function for dilation
def dilation(img, SE):
  imgDilate= cv.dilate(img,SE,1)
  return imgDilate
SE= cv.getStructuringElement(cv.MORPH_RECT,(3,3)) 
AeB= erosion(img,SE)
AoBdB= dilation(AeB,SE) 

plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(AoBdB), plt.title('Citra Hasil Opening')
```
